```python
import seaborn as sns
```

```python
print(" ")
sns.jointplot(data=dataset,x='Price',y='Aggregate rating',kind='reg')
plt.show()
```

```python
print("\nBased on Aggregate ratings classifying the customers Feedbacks!\n ")
fig = px.pie(dataset, values='Feedbacks', names='Aggregate rating')
fig.show()
```

```python
print("\n Resevations in Each restaurents!\n")
fig=px.bar(dataset.head(),x='Restaurent Name',y='Reservation',template='ggplot2')
fig.show()
```

```python
print("\n Item Prices and based on Restaurent Names!\n")
fig=px.bar(dataset.head(100),x='Restaurent Name',y='Price',template='ggplot2')
fig.show()
```

```python
print("\n predicting the online ratings !, Text !, And Customers Feedbacks! \n")
fig = px.line(dataset, x="Aggregate rating", y="Text", color='Feedbacks',markers=True,color_discrete_sequence=['White','orange'],template='plotly_dark')
fig.show()
```
